## Question 1: Difference between descriptive and inferential statistics

-        Descriptive statistics summarizes data using mean, median, mode, tables, and graphs.
-        Inferential statistics draws conclusions about a population using sample data.
-        Example: Class average (descriptive) and predicting population income (inferential).

---

## Question 2: Sampling, random sampling, and stratified sampling

-        Sampling is selecting a subset of a population for analysis.
-        Random sampling gives equal chance to all individuals.
-        Stratified sampling divides population into groups and samples from each.

---

## Question 3: Mean, median, and mode

-        Mean is the average of values.
-        Median is the middle value.
-        Mode is the most frequent value.
-        They represent the central tendency of data.

---

## Question 4: Skewness and kurtosis

-        Skewness shows data asymmetry.
-        Kurtosis shows data peakedness.
-        Positive skew means a long right tail.

```python
# Question 5: Mean, Median, Mode
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
print("Mean:", statistics.mean(numbers))
print("Median:", statistics.median(numbers))
print("Mode:", statistics.mode(numbers))

# Output:
# Mean: 19
# Median: 19
# Mode: 12
```

```python
# Question 6: Covariance and Correlation
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

print("Covariance:", np.cov(list_x, list_y)[0][1])
print("Correlation:", np.corrcoef(list_x, list_y)[0][1])

# Output:
# Covariance: 312.5
# Correlation: 0.994
```

```python
# Question 7: Boxplot and Outlier
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
plt.boxplot(data)
plt.title("Boxplot")
plt.show()
```

```python
# Question 8: Correlation between Advertising Spend and Daily Sales
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

print("Correlation:", np.corrcoef(advertising_spend, daily_sales)[0][1])

```

```python
# Question 9: Histogram for Survey Scores
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores)
plt.title("Survey Histogram")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()

# Output:
# Histogram displayed showing distribution of survey scores
```
